```python
import os
import cv2
import matplotlib.pyplot as plt
data = []
size = 100
categories = ["angry","disgust","fear","happy","neutral","sad","surprise"]
for i in categories:
  folder_path = os.path.join("/content/dataset/dataset",i)
  cat_index = categories.index(i)
  for img in os.listdir(folder_path):
    abs_img_path = os.path.join(folder_path,img)
    img_arr = cv2.imread(abs_img_path)
    img_arr = cv2.resize(img_arr,[size,size])
    data.append([img_arr,cat_index])
```

```python
random.shuffle(data)
```

Feature extraction

```python
X = []
Y = []
for features,label in data:
  X.append(features)
  Y.append(label)
```

Standarizing the image

```python
X = X/225
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=22,shuffle=True)
```

```python

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator

model = keras.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(size, size, 3)),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(7, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))
```

```python

loss, accuracy = model.evaluate(X_test, y_test)
print("Loss:", loss)
print("Accuracy:", accuracy)
```

# ACCURACY OF 98%
Which means the dataset able to successfully detect 98% variance in emotion within data

```python


import matplotlib.pyplot as plt
import numpy as np

random_index = random.randint(0, len(X_test) - 1)
test_image = X_test[random_index]

prediction = model.predict(np.expand_dims(test_image, axis=0))

predicted_class_index = np.argmax(prediction)

predicted_class_name = categories[predicted_class_index]

print("Predicted class:", predicted_class_name)

plt.imshow(test_image)
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np
def predict_emotion(path):

  img_arr = cv2.imread(path)
  img_arr = cv2.resize(img_arr, [size, size])
  img_arr = np.array(img_arr)
  img_arr = img_arr / 225
  prediction = model.predict(np.expand_dims(img_arr, axis=0))
  predicted_class_index = np.argmax(prediction)

  predicted_class_name = categories[predicted_class_index]
  fig,(ax1,ax2) = plt.subplots(1,2)
  ax1.imshow(cv2.imread(path))
  ax1.set_title("Original image")
  ax2.imshow(img_arr)
  ax2.set_title("Processed 100X100")
  plt.show()
  return predicted_class_name, img_arr
```
